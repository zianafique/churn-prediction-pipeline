```python
import pandas as pd
import pickle

# Load original
df_original = pd.read_csv('../data/raw/WA_Fn-UseC_-Telco-Customer-Churn.csv')

# Load processed
X_train = pd.read_csv('../data/processed/X_train.csv')
y_train = pd.read_csv('../data/processed/y_train.csv')

print("Original data:")
print(f"  Shape: {df_original.shape}")
print(f"  First row:\n{df_original.iloc[0]}\n")

print("Processed data:")
print(f"  Shape: {X_train.shape}")
print(f"  First row:\n{X_train.iloc[0]}\n")

print("Data types changed:")
print(f"  Original:\n{df_original.dtypes}")
print(f"\n  Processed:\n{X_train.dtypes}")
```

```python
# Load encoders
with open('../data/processed/encoders.pkl', 'rb') as f:
    encoders = pickle.load(f)

print("Categorical encodings:")
for col, encoder in encoders.items():
    encoding_map = dict(zip(encoder.classes_, encoder.transform(encoder.classes_)))
    print(f"\n{col}:")
    print(f"  {encoding_map}")
```

```python
# Load scaler
with open('../data/processed/scaler.pkl', 'rb') as f:
    scaler = pickle.load(f)

print("Scaler statistics:")
print(f"  Means: {scaler.mean_}")
print(f"  Stds: {scaler.scale_}")
print(f"\nAfter scaling, all features have:")
print(f"  Mean ≈ 0")
print(f"  Std ≈ 1")

# Show actual values
print(f"\nActual scaled values (first customer):")
print(f"  {X_train.iloc[0].head(10).values}")
```

```python
print("Churn distribution in all sets:")
print(f"\nOriginal (full data):")
print(f"  {y_train.value_counts(normalize=True)}")

y_val = pd.read_csv('../data/processed/y_val.csv')
y_test = pd.read_csv('../data/processed/y_test.csv')

print(f"\nTrain set:")
print(f"  {y_train.value_counts(normalize=True)}")

print(f"\nVal set:")
print(f"  {y_val.value_counts(normalize=True)}")

print(f"\nTest set:")
print(f"  {y_test.value_counts(normalize=True)}")

print(f"\n✅ All sets have same ~26.5% churn rate (stratified successfully!)")
```
